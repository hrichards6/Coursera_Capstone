Convert Categorical features to numerical values

```python
Feature = collisions_data[['WEATHER','ROADCOND','LIGHTCOND','COLLISIONTYPE','dayofweek','weekend']]
Feature = pd.concat([Feature,pd.get_dummies(collisions_data[['WEATHER','ROADCOND','LIGHTCOND','COLLISIONTYPE', 'dayofweek','weekend']])], axis=1)
Feature.drop(['WEATHER', 'ROADCOND','LIGHTCOND','COLLISIONTYPE','dayofweek','weekend'], axis = 1,inplace=True)
Feature.head()
```

```python
X = Feature
X[0:5]
```

```python
y = collisions_data['SEVERITYCODE'].replace([1,2],['PROPDAMAGE','INJURY']).values
y[0:100]
```

```python
from sklearn import preprocessing
```

```python

X= preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

K nearest neighbour

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 4)
print('Train set: ', X_train.shape, y_train.shape )
print( 'Test set: ', X_test.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
#Training
#Calculate accuracy of KNN for different values of K
Ks = 10
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
confusionMx = [];
for n in range(1,Ks):
    #Train model and predict
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train, y_train)
    yhat = neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)
    
    std_acc[n-1] = np.std(yhat == y_test)/np.sqrt(yhat.shape[0])
    
mean_acc
```

```python
#plot model accuracy for different numbers of neighbors
plt.plot(range(1,Ks),mean_acc, 'g')
plt.fill_between(range(1,Ks),mean_acc - 1*std_acc, mean_acc + 1*std_acc, alpha = 0.10)
plt.legend(('Accuracy', '+/-3xstd'))
plt.ylabel('Accuracy')
plt.xlabel('Value of K')
plt.tight_layout()
plt.show()
```

```python
print('The best accuracy was with: ', mean_acc.max(), 'with k= ', mean_acc.argmax()+1)
```
