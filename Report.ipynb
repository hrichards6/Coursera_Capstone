### Introduction/ Business Case

The Seattle Police Department have provided us data on all recorded collisions from 2004 to present. I will be using this dataset for this capstone project to build a machine learning model used to predict the severity of an accident dependent on factors such as environmental conditions (weather, road surface and lighting conditions), irresponsible driving (speeding, distracted and under the influence) and also what time of day/day of week it is.
This machine learning model will be able to predict the severity of a collision, given other factors. With this information motorists will now be able to make an educated decision on whether or not to venture out onto the roads. This information could also be used to give alerts on the radio so that motorists are warned if conditions mean there are higher chances of severe crashes, for example, if it is snowing and there is black ice on the road.
